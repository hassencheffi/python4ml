**Seaborn** est une bibliothèque de **visualisation de données** basée sur **Matplotlib**. Elle fournit une interface de haut niveau pour créer des graphiques attractifs. Dans ce tutoriel, nous explorerons quelques exemples d'utilisation de Seaborn.

### Chargement d'un jeu de données

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

tips.head(20)
tips.describe()
```

```python
total_bill_sum_by_sex = tips.groupby("day")["total_bill"].sum().reset_index()
print(total_bill_sum_by_sex)

# Créer un graphique en barres avec Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(data=total_bill_sum_by_sex, x="day", y="total_bill", palette="Set2")
plt.title("Histogramme du montant total de la facture")
plt.xlabel("Montant total de la facture")
plt.ylabel("Fréquence")
plt.show()
```

### Exemple 1 : Histogramme

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Histogramme du montant total de la facture")
plt.xlabel("Montant total de la facture")
plt.ylabel("Fréquence")
plt.show()
```

### Exemple 2 : Nuage de points avec régression linéaire

```python
sns.lmplot(data=tips, x="total_bill", y="tip")
plt.title("Nuage de points avec régression linéaire")
plt.xlabel("Montant total de la facture")
plt.ylabel("Pourboire")
plt.show()
```

### Exemple 3 : Carte thermique (heatmap) de la matrice de corrélation

```python
correlation_matrix = tips.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Matrice de corrélation")
plt.show()
```

### Exemple 4 : Boîte à moustaches (boxplot) des montants de facture par jour

```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Boîte à moustaches des montants de facture par jour")
plt.xlabel("Jour de la semaine")
plt.ylabel("Montant total de la facture")
plt.show()
```
